```python
def rotation(string):
    '''Создаёт матрицу со сдвигами нашей строки'''
    matrix = []
    for i in range(len(string)):
        matrix.append(string[-(i+1):] + string[:-(i+1)])
    return (matrix)
print(rotation("ABAABA$"))
```

```python
def columns(matrix):
    """Возвращает первый и последний вектора нашей вращающейся матрицы"""
    column_last = ''
    column_first = ''
    for i in range(len(matrix)):
        column_last += matrix[i][-1]
        column_first += matrix[i][0]
    return column_first, column_last
```

```python
def bwt(string: str):
    """Делает BWT преобразование строки
    Возвращает начальный и конечный столбец"""
    string += '$'
    string_matrix = sorted(rotation(string))
    first_column, last_column= columns(string_matrix)
    
    return first_column, last_column
print(bwt("ABAABA"))
```

```python
def ranking_vector(vector) -> list :
    """Нумерует символы в порядке вхождения"""
    counter = {}
    ranked_vector = []
    for simbol in vector:
        if simbol in counter:
            counter[simbol] += 1
            ranked_vector.append((simbol, counter[simbol]))
        else:
            counter[simbol] = 0
            ranked_vector.append((simbol, counter[simbol]))
    return ranked_vector
print(ranking_vector(bwt("ABAABA")[0]),ranking_vector(bwt("ABAABA")[1]))  #нумеруем независимо, тк порядок элементов сохраняется при BWT-преобразовании
```

```python
def make_string(F,L) -> list :
    """сборка исходной строки по первому и последнему столбцам матрицы сдвигов"""
    
    start=F[L.index(('$',0))] #тк доллар в последнем столбце является предшественником нулевого элемента исходной строки
    string=[start]
    while start[0]!='$':
        start=F[L.index(start)] #тк элемент последнего столбца явл-ся предшественником соответсвующего эл-та первого столбца
        string.append(start)
    return string
        
    
make_string(ranking_vector(bwt("ABAABA")[0]),ranking_vector(bwt("ABAABA")[1]))
```

```python
def search_pattern(sequence,pattern:str):
    '''Ищем подстроку в строке с использованием BWT'''
    ends=[]
    F, L=bwt(sequence)
    F=ranking_vector(F)
    L=ranking_vector(L)
    pattern+='#'
    for l in L:
        if l[0]==pattern[0]:
            start=l
            for elem in pattern[1:]:
                if elem=='#':
                    ends.append(start)
                    break
                start=F[L.index(start)]
                if elem!=start[0]:
                    break
    if ends==[]:
        return -1  #озрначает отсутствие вхождений
    origin=make_string(F,L) #в этой строчке индексация будет соответствовать не исходной, а той, что получилась при индкесации BWT-колонок
    result=[]
    for elem in ends:
        result.append(origin.index(elem)-(len(pattern)-1)+1)  #если индексировать строку с единицы, то должно быть + два
    return result
    
search_pattern("ABAABA","ABA") 
```
